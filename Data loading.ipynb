```python
import os
import shutil
import glob
import random

# glob.glob returns a list of pathnames which are of the form *.jpg
daisy = glob.glob(os.path.join('/home/sayan/Desktop/Flower Classification/flowers/daisy', '*.jpg'))
dandelion = glob.glob(os.path.join('/home/sayan/Desktop/Flower Classification/flowers/dandelion', '*.jpg'))
rose = glob.glob(os.path.join('/home/sayan/Desktop/Flower Classification/flowers/rose', '*.jpg'))
sunflower = glob.glob(os.path.join('/home/sayan/Desktop/Flower Classification/flowers/sunflower', '*.jpg'))
tulip = glob.glob(os.path.join('/home/sayan/Desktop/Flower Classification/flowers/tulip', '*.jpg'))

random.shuffle(daisy);
random.shuffle(dandelion);
random.shuffle(rose);
random.shuffle(sunflower);
random.shuffle(tulip);

daisytrain = daisy[:669];
daisytest = daisy[669:];

dandeliontrain = dandelion[:669];
dandeliontest = dandelion[669:];

rosetrain = rose[:669];
rosetest = rose[669:];

sunflowertrain = sunflower[:669];
sunflowertest = sunflower[669:];

tuliptrain = tulip[:669];
tuliptest = tulip[669:];

# Making directory using os module
os.makedirs('/home/sayan/Desktop/Flower Classification/train/daisy');
os.makedirs('/home/sayan/Desktop/Flower Classification/test/daisy');

os.makedirs('/home/sayan/Desktop/Flower Classification/train/dandelion');
os.makedirs('/home/sayan/Desktop/Flower Classification/test/dandelion');

os.makedirs('/home/sayan/Desktop/Flower Classification/train/rose');
os.makedirs('/home/sayan/Desktop/Flower Classification/test/rose');

os.makedirs('/home/sayan/Desktop/Flower Classification/train/sunflower');
os.makedirs('/home/sayan/Desktop/Flower Classification/test/sunflower');

os.makedirs('/home/sayan/Desktop/Flower Classification/train/tulip');
os.makedirs('/home/sayan/Desktop/Flower Classification/test/tulip');

for f in daisytrain:
    shutil.copy(f, '/home/sayan/Desktop/Flower Classification/train/daisy/');
for f in daisytest:
    shutil.copy(f, '/home/sayan/Desktop/Flower Classification/test/daisy/');

for f in dandeliontrain:
    shutil.copy(f, '/home/sayan/Desktop/Flower Classification/train/dandelion/');
for f in dandeliontest:
    shutil.copy(f, '/home/sayan/Desktop/Flower Classification/test/dandelion/');

for f in rosetrain:
    shutil.copy(f, '/home/sayan/Desktop/Flower Classification/train/rose/');
for f in rosetest:
    shutil.copy(f, '/home/sayan/Desktop/Flower Classification/test/rose/');
    
for f in sunflowertrain:
    shutil.copy(f, '/home/sayan/Desktop/Flower Classification/train/sunflower/');
for f in sunflowertest:
    shutil.copy(f, '/home/sayan/Desktop/Flower Classification/test/sunflower/');
    
for f in tuliptrain:
    shutil.copy(f, '/home/sayan/Desktop/Flower Classification/train/tulip/');
for f in tuliptest:
    shutil.copy(f, '/home/sayan/Desktop/Flower Classification/test/tulip/');
```
